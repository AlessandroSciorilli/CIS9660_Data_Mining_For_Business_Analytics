```python
#Basic Plots in Matplotlib

import matplotlib.pyplot as plt

#Source - https://matplotlib.org/stable/index.html
```

```python
#Plotting a Basic Line Plot

day = [1, 2, 3, 4, 5, 6, 7]
temperature = [32.3, 35, 22, 45, 50, 28, 40]

plt.plot(day, temperature)
plt.title('New York Temperature', fontsize = 20)
plt.axis([0, 5, 0, 60]) #set the limits on the different axes in the format [x_min, x_max, y_min, y_max]
plt.xlabel('Day', fontsize = 15)
plt.ylabel('Temperature', fontsize = 15)
#setting the size of the figure
plt.rcParams['figure.figsize'] = [15 , 6]

plt.show()
```

```python
#Changing the color of the line

day = [1, 2, 3, 4, 5, 6, 7]

temperature = [32.3, 35, 22, 45, 50, 28, 40]
#color parameter - change to our preference 

plt.plot(day, temperature, color ='g')
plt.title('New York City Temperature', fontsize = 20)
plt.axis([0, 5, 0, 60]) #set the limits on the different axes in the format [x_min, x_max, y_min, y_max]
plt.xlabel('Day', fontsize = 10)
plt.ylabel('Temperature', fontsize = 10)
#setting the size of the figure
plt.rcParams['figure.figsize'] = [10,4]

plt.show()
```

```python
#Adding the the data points to the graph as a marker


day = [1, 2, 3, 4, 5, 6, 7]
temperature = [32.3, 35, 22, 45, 50, 28, 40]

#Note the change of color in the line and the marker parameter
plt.plot(day, temperature, color ='g', marker='o')
plt.title('New York City Temperature', fontsize = 20)
plt.axis([0, 5, 0, 60]) #set the limits on the different axes in the format [x_min, x_max, y_min, y_max]
plt.xlabel('Day', fontsize = 10)
plt.ylabel('Temperature', fontsize = 10)
#setting the size of the figure
plt.rcParams['figure.figsize'] = [10,4]

plt.show()
```

```python
#Line Graphs with Categorical Data
teams = ['Giants', 'Jets', 'Ravens', 'Rams', 'Bears', 'Eagles']
points = [50, 30, 45, 22, 8, 35]
plt.title('Football Teams vs Points')
plt.xlabel('Teams')
plt.ylabel('Points')
plt.plot(teams,points)
plt.show()
```

```python
#Rotating the x labels - useful if you have labels with longer names
teams = ['Giants', 'Jets', 'Ravens', 'Rams', 'Bears', 'Eagles']
points = [50, 30, 45, 22, 8, 35]
plt.title('Football Teams vs Points')
plt.xlabel('Teams')
#Note the use of the rotation parameter which is set to 90 degrees
plt.xticks(rotation = 90)
plt.ylabel('Points')
plt.plot(teams,points)
plt.show()
```

```python
#Plotting multiple lines on a single graph 

dates= ['1/7/22', '1/14/22', '1/21/22', '1/28/22', '2/4/22']
opening_price = [40, 20, 45, 48, 30]
closing_price = [38, 10, 45, 50, 41]

plt.title('Opening vs Closing Stock Prices for a Company')
#Note below the various plots - same x values but different y values

plt.plot(dates, opening_price, color='b', label = 'opening price', linewidth= 3)

plt.plot(dates, closing_price, color='r', label = 'closing price', linewidth=3)

#Note the location of the legend - try changing the value of loc to 1,2,3, etc. What do you notice?
plt.legend(loc=4)
plt.show()
```

```python
#Adding a grid to the background of our plot
#Plotting multiple lines on a single graph 

#Note the new imports
from matplotlib import style

#setting the parameter us a grid plot
style.use('ggplot')

dates= ['1/7/22', '1/14/22', '1/21/22', '1/28/22', '2/4/22']
opening_price = [40, 20, 45, 48, 30]
closing_price = [38, 10, 45, 50, 41]

plt.title('Opening vs Closing Stock Prices for a Company')
#Note below the various plots - same x values but different y values
plt.plot(dates, opening_price, color='b', label = 'opening price', linewidth= 3)
plt.plot(dates, closing_price, color='r', label = 'closing price', linewidth=3)

#Note the location of the legend - try changing the value of loc to 1,2,3, etc. What do you notice?
plt.legend(loc=4)

#Giving the plot a grid-like background
plt.grid(True,color='c')
plt.show()
```

```python
#Creating subplots

#our data
teams = ['Giants', 'Jets', 'Ravens', 'Rams', 'Bears', 'Eagles']
points = [50, 30, 45, 22, 8, 35]

#creating the layout for our subplots - one row with 2 figures
figure, axes = plt.subplots(nrows=1, ncols=2)

plt.figure(figsize=(20,3))

#We have create a one dimensional array of plots
#Index 0 will have the bar plot
axes[0].bar(teams, points)

#index 1 will have the line plot
axes[1].plot(teams, points)
figure.suptitle('Teams vs Points')
plt.show()
```

```python
#Creating more subplots

#our data
teams = ['Giants', 'Jets', 'Ravens', 'Rams', 'Bears', 'Eagles']
points = [50, 30, 45, 22, 8, 35]


dates= ['1/7/22', '1/14/22', '1/21/22', '1/28/22', '2/4/22']
opening_price = [40, 20, 45, 48, 30]

#creating the layout for our subplots - 2 rows and 2 columns with 4 figures
figure, axes = plt.subplots(nrows=2, ncols=2)

plt.figure(figsize=(20,3))

#We have create a one dimensional array of plots
#Index 0,0 will have the bar plot
axes[0,0].bar(teams, points)
#index 0,1 will have the line plot
axes[0,1].plot(teams, points)

#Index 1,0 will have the scatter plot
axes[1,0].scatter(teams, points)

#index 1,1 will have the line plot of the stock data
axes[1,1].plot(dates, opening_price)

figure.suptitle('An Example of 4 Subplots Within a Single Figure')
plt.show()
```
