#Titanic Dataset
The Titanic dataset is a consists of passengers on the Titanic, a British passenger liner that sank in the North Atlantic Ocean in 1912, killing 1,517 people. The dataset contains information about the passengers, such as their names, ages, genders, and cabin numbers. It also contains information about whether or not they survived the sinking.




The following is a description of the columns in the Titanic dataset:

* PassengerId: A unique identifier for each
passenger.
* Survived: A binary variable indicating whether or not the passenger survived the sinking.
* Pclass: The passenger's class, with 1 being first class, 2 being second class, and 3 being third class.
* Name: The passenger's name.
* Sex: The passenger's sex.
* Age: The passenger's age in years.
* SibSp: The number of siblings or spouses the passenger had on board.
* Parch: The number of parents or children the passenger had on board.
* Ticket: The passenger's ticket number.
* Fare: The passenger's fare.
* Cabin: The passenger's cabin number.
* Embarked: The port where the passenger embarked, with C being Cherbourg, Q being Queenstown, and S being Southampton.

Importing Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

Load the Titanic dataset into a pandas dataframe

```python
titanic_data = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

Display the first few rows of the dataset

```python
print(titanic_data.head())
```

Plot 1: Bar plot of passenger survival


```python
sns.countplot(x='Survived', data=titanic_data)
plt.title('Passenger Survival')
plt.show()
```

Plot 2: Bar plot of passenger Pclass distribution

```python
sns.countplot(x='Pclass', data=titanic_data)
plt.title('Passenger Pclass Distribution')
plt.show()
```

Plot 3: Box plot of passenger Ages by Pclass

```python
sns.boxplot(x='Pclass', y='Age', data=titanic_data)
plt.title('Passenger Ages by Pclass')
plt.show()
```

Plot 4: Distribution of passenger Fares

```python
sns.histplot(x='Fare', data=titanic_data)
plt.title('Distribution of Passenger Fares')
plt.show()
```

Plot 5: Swarm plot of passenger Ages by Sex

```python
sns.swarmplot(x='Sex', y='Age', data=titanic_data)
plt.title('Passenger Ages by Sex')
plt.show()
```

Plot 6: Count plot of passenger embarkation port

```python

sns.countplot(x='Embarked', data=titanic_data)
plt.title('Passenger Embarkation Port')
plt.show()
```

Plot 7: Bar plot of passenger survival by Pclass

```python
sns.countplot(x='Pclass', hue='Survived', data=titanic_data)
plt.title('Passenger Survival by Pclass')
plt.show()
```

Plot 8: Violin plot of passenger Ages by Pclass


```python
sns.violinplot(x='Pclass', y='Age', data=titanic_data)
plt.title('Passenger Ages by Class')
plt.show()
```

Plot 9: Distribution of passenger Ages


```python
sns.histplot(x='Age', data=titanic_data)
plt.title('Distribution of Passenger Ages')
plt.show()
```

Plot 10: Box plot of passenger Ages by survival


```python
sns.boxplot(x='Survived', y='Age', data=titanic_data)
plt.title('Passenger Ages by Survival')
plt.show()
```

Plot 11: Count plot of passenger survival by Sex


```python
sns.countplot(x='Sex', hue='Survived', data=titanic_data)
plt.title('Passenger Survival by Sex')
plt.show()
```

Plot 12: Scatter plot of passenger Ages vs. Fares


```python
sns.scatterplot(x='Age', y='Fare', data=titanic_data)
plt.title('Passenger Ages vs. Fares')
plt.show()
```

Plot 13: Heatmap of correlations between numerical features


```python
corr_matrix = titanic_data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Between Numerical Features')
plt.show()
```

Plot 14: Bar plot of passenger survival by Age group

```python
titanic_data['Age_group'] = pd.cut(titanic_data['Age'], bins=[0, 18, 30, 50, 100], labels=['0-18', '19-30', '31-50', '51+'])
sns.countplot(x='Age_group', hue='Survived', data=titanic_data)
plt.title('Passenger Survival by Age Group')
plt.show()
```

Plot 15: Count plot of passenger survival by Port Embarked


```python
titanic_data['Embarked'] = titanic_data['Embarked']
sns.countplot(x='Embarked', hue='Survived', data=titanic_data)
plt.title('Passenger Survival by Embarked')
plt.show()
```

Plot 16: Stacked bar plot of passenger survival by Pclass and Sex

```python
sns.pairplot(titanic_data.dropna(), vars=['Age', 'Fare', 'Pclass'], hue='Survived', markers=['o', 's'])
plt.suptitle('Pairwise Scatter Plot of Numerical Features')
plt.show()
```
